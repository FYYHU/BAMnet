```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
CT_df = pd.read_csv('./DataCT_10000_0.05_0.02_1_80_30_10exp.csv')
R_df = pd.read_csv('./DataR_10000_0.05_0.02_1_80_30_10exp.csv')
MF_df = pd.read_csv('./DataMF_10000_0.05_0.02_1_80_30_10exp.csv')
```

```python
CT_df_grouped = CT_df.groupby('time').mean()
R_df_grouped = R_df.groupby('time').mean()
MF_df_grouped = MF_df.groupby('time').mean()
```

```python
x_axis = CT_df_grouped.index
CTy_S_axis = CT_df_grouped['S']
CTy_I_axis = CT_df_grouped['I']
CTy_R_axis = CT_df_grouped['R']
Ry_S_axis = R_df_grouped ['S']
Ry_I_axis = R_df_grouped ['I']
Ry_R_axis = R_df_grouped ['R']
MFy_S_axis = MF_df_grouped ['S']
MFy_I_axis = MF_df_grouped ['I']
MFy_R_axis = MF_df_grouped ['R']

plt.plot(x_axis, CTy_I_axis, label='CT I')
plt.plot(x_axis, Ry_I_axis, label='R I')
plt.plot(x_axis, MFy_I_axis, label='MF I')
plt.xlim(40, 80)
plt.legend()
```
